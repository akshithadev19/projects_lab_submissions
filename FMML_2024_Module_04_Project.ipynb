## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

### Example 3 - Divergence and Overflow

#### $$h(x) = x^5 - 2x^4 + 2$$

````python
## TASK-4 : Code the same things as they were done for the first two examples
````

````python
###task 4A:
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative for h(x)
def h(x):
    return x**5 - 2*x**4 + 2

def dh(x):
    return 5*x**4 - 8*x**3  # Derivative of h(x)

# Gradient Descent Function
def gradient_descent(derivative_func, initial_guess, multiplier=0.0001, precision=0.0001, max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        # Clamp the value of x to avoid overflow
        new_x = np.clip(new_x, -10, 10)  # Clamping x between -10 and 10

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            print(f'Converged in {n} iterations.')
            break
    return new_x, x_list, slope_list

# Perform Gradient Descent for multiple initial guesses
initial_guesses = [-1.5, 0, 1.5]  # Reduced initial guesses to avoid large numbers
results = []

for initial_guess in initial_guesses:
    local_min, list_x, deriv_list = gradient_descent(dh, initial_guess)
    results.append((initial_guess, local_min, list_x, deriv_list))
    print(f"Initial guess: {initial_guess}, Local minimum: {local_min}, Function value at min: {h(local_min)}")

# Visualize the descent for all initial guesses
x_2 = np.linspace(-2, 2, 1000)

for result in results:
    initial_guess, local_min, list_x, deriv_list = result

    plt.figure(figsize=[15, 5])

    # Chart 1: Cost function h(x)
    plt.subplot(1, 2, 1)
    plt.xlim(-2, 2)
    plt.ylim(-4, 10)
    plt.title('Cost function h(x)', fontsize=17)
    plt.xlabel('X', fontsize=16)
    plt.ylabel('h(x)', fontsize=16)
    plt.plot(x_2, h(x_2), color='blue', linewidth=3)
    plt.scatter(list_x, h(np.array(list_x)), color='red', s=80, alpha=0.6, label='Descent Steps')
    plt.legend()

    # Chart 2: Derivative dh(x)
    plt.subplot(1, 2, 2)
    plt.xlim(-2, 2)
    plt.ylim(-10, 15)
    plt.title('Slope of h(x)', fontsize=17)
    plt.xlabel('X', fontsize=16)
    plt.ylabel('dh(x)', fontsize=16)
    plt.plot(x_2, dh(x_2), color='orange', linewidth=3)
    plt.scatter(list_x, deriv_list, color='red', s=80, alpha=0.6, label='Slope Descent')
    plt.legend()
    plt.grid()

    plt.show()
````

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

````python
## TASK-8 : Code the same things as they were done for the first two examples
import numpy as np
import matplotlib.pyplot as plt

def f1(x, y):
    return 13 - x**2 - y**2 + 1

def df1(x, y):
    df_dx = -2 * x
    df_dy = -2 * y
    return df_dx, df_dy

def f2(x, y):
    r = 3 - x**2 - y**2
    return 1 / r + 1

def df2(x, y):
    r = 3 - x**2 - y**2
    df_dx = (2 * x) / (r**2)
    df_dy = (2 * y) / (r**2)
    return df_dx, df_dy


def gradient_descent(derivative_func, initial_guess, multiplier=0.1, precision=0.0001, max_iter=500):
    x, y = initial_guess
    x_list, y_list, cost_list = [x], [y], [f1(x, y) if derivative_func == df1 else f2(x, y)]

    for n in range(max_iter):
        df_dx, df_dy = derivative_func(x, y)
        x -= multiplier * df_dx
        y -= multiplier * df_dy

        step_size = np.sqrt(df_dx**2 + df_dy**2)
        x_list.append(x)
        y_list.append(y)
        cost_list.append(f1(x, y) if derivative_func == df1 else f2(x, y))

        if step_size < precision:
            print(f"Converged in {n} iterations.")
            break

    return x, y, x_list, y_list, cost_list

initial_guess1 = (2, 2)
local_min1, local_min2, x_list1, y_list1, cost_list1 = gradient_descent(df1, initial_guess1, multiplier=0.1)

initial_guess2 = (1, 1)  # Starting point for f2
local_min2, local_min2, x_list2, y_list2, cost_list2 = gradient_descent(df2, initial_guess2, multiplier=0.1)

# Plot the reduction in cost for each iteration (TASK-8)
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.plot(cost_list1, label="Cost function (f1)", color='blue')
plt.title('Cost Reduction for f1(x, y) = 13 - x^2 - y^2 + 1', fontsize=14)
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Cost', fontsize=12)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(cost_list2, label="Cost function (f2)", color='green')
plt.title('Cost Reduction for f2(x, y) = 1 / r + 1', fontsize=14)
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Cost', fontsize=12)
plt.legend()

plt.tight_layout()
plt.show()

# Printing the results
print("Function 1:")
print(f"Local minimum for f1: ({local_min1}, {local_min2})")
print(f"Number of iterations for f1: {len(x_list1)}")

print("\nFunction 2:")
print(f"Local minimum for f2: ({local_min1}, {local_min2})")
print(f"Number of iterations for f2: {len(x_list2)}")
````

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````

````python
import numpy as np
import matplotlib.pyplot as plt

def mse(y, y_hat):
    return np.mean((y - y_hat) ** 2)

def grad(x, y, thetas):
    n = y.size

    theta0_slope = (-2/n) * np.sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * np.sum((y - thetas[0] - thetas[1]*x) * x)
    return np.array([theta0_slope, theta1_slope])


multiplier = 0.01
thetas = np.array([2.9, 2.9])

x_5 = np.array([1, 2, 3, 4, 5, 6, 7])
y_5 = np.array([1.1, 1.9, 3.0, 4.1, 5.1, 6.0, 7.1])

plot_vals = thetas.reshape(1, 2)
mse_vals = [mse(y_5, thetas[0] + thetas[1] * x_5)]


for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)


    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)


    mse_vals.append(mse(y_5, thetas[0] + thetas[1] * x_5))

print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('Final MSE is:', mse_vals[-1])

plt.plot(mse_vals)
plt.xlabel('Iteration')
plt.ylabel('MSE')
plt.title('MSE Reduction over Iterations')
plt.show()
````
